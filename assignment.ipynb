```python
# Q1. What is the role of feature selection in anomaly detection?

Feature selection helps identify the most relevant features, reducing noise and dimensionality, improving model accuracy, interpretability, and 
computational efficiency in anomaly detection.

# Q2. What are some common evaluation metrics for anomaly detection algorithms and how are they computed?

Precision: TP/(TP+FP)
Recall (Sensitivity): TP/(TP+FN)
F1-Score: 2 X (Precision X Recall)/(Precision+Recall)
AUC-ROC: Measures the trade-off between true positive rate and false positive rate.
Confusion Matrix: Tracks TP, TN, FP, and FN for binary classification tasks

# Q3. What is DBSCAN and how does it work for clustering?

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) clusters data points based on density. It groups points within a specified 
radius (ϵ) and a minimum number of points (MinPts), marking others as noise or border points.

# Q4. How does the epsilon parameter affect the performance of DBSCAN in detecting anomalies?

Smaller ϵ: Identifies finer clusters, increasing noise points and anomalies.
Larger ϵ: Merges clusters, reducing sensitivity to anomalies but increasing noise misclassification.

# Q5. What are the differences between the core, border, and noise points in DBSCAN, and how do they relate to anomaly detection?

Core points: Have at least MinPts neighbors within ϵ.
Border points: Lie within ϵ of a core point but lack enough neighbors to be core points.
Noise points: Do not belong to any cluster, often treated as anomalies.

# Q6. How does DBSCAN detect anomalies and what are the key parameters involved in the process?

DBSCAN treats noise points as anomalies by detecting points that do not belong to any dense region. Key parameters:
Epsilon (ϵ): Radius for neighborhood density.
MinPts: Minimum points required in a neighborhood to form a cluster.

# Q7. What is the make_circles package in scikit-learn used for?

The make_circles function generates synthetic datasets of concentric circles, often used to test clustering and classification algorithms' ability 
to separate non-linear patterns.

# Q8. What are local outliers and global outliers, and how do they differ from each other?

Local outliers: Deviate significantly within a local region compared to neighbors (e.g., LOF).
Global outliers: Deviate significantly from the entire dataset (e.g., Isolation Forest).

# Q9. How can local outliers be detected using the Local Outlier Factor (LOF) algorithm?

LOF compares the density of a data point with its neighbors. Points with significantly lower density than their neighbors are flagged as local outliers.

# Q10. How can global outliers be detected using the Isolation Forest algorithm?

Isolation Forest isolates data points by recursively partitioning features. Points requiring fewer partitions are considered global outliers.

# Q11. What are some real-world applications where local outlier detection is more appropriate than global outlier detection, and vice versa?

Local Outlier Detection: Fraud detection in localized regions, network traffic analysis.
Global Outlier Detection: Industrial equipment failure detection, rare event detection in large-scale datasets.
```
